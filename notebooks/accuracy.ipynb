```python
#facilities
import numpy as np
from matplotlib import pyplot as plt
from scipy.interpolate import make_interp_spline
plt.figure(figsize=(8,5))
plt.xlabel("Epoch",fontsize=12)
plt.ylabel("Acc@5 accuracy",fontsize=12)
x = np.array([0,10,20,30,40,50,60,70])
y = np.array([0.86, 0.945, 0.95, 0.98, 0.98, 0.98, 0.98 ,0.975])
y1 = np.array([0.325, 0.935, 0.985, 0.985, 0.985, 0.99, 0.985, 0.99])
x_smooth = np.linspace(x.min(), x.max(), 300)
y_smooth = make_interp_spline(x, y)(x_smooth)
y1_smooth = make_interp_spline(x, y1)(x_smooth)
plt.tick_params(labelsize=12)
plt.plot(x_smooth, y_smooth, label='Lodi')
plt.plot(x_smooth, y1_smooth, linestyle='--',label='GraphConv')
plt.legend()
```
